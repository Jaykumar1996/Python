```python
# Assignment 3 Problem 1
# Find the highest daily gain and its date, the highest daily loss and its date
# By Michael Levkovitch, Jacob Aufdenberg, Prabesh Amatya, and Jay Patel



import numpy as np
original_array = np.genfromtxt('SP500.csv', delimiter=',', dtype=None, encoding="utf8")

b = original_array[1:,0]
a = original_array[1:,1:].astype(float)


    
max_row = 0
min_row = 0
    

for i in range(len(a)):
    if (a[i,3]-a[i,0] > a[max_row,3]-a[max_row,0]):
        max_row = i
    if (a[i,3]-a[i,0] < a[min_row,3]-a[min_row,0]):
        min_row = i
    
print('The highest daily loss in stock market history occured on {0}'.format(b[min_row]))
print('On that day, the stock market dropped {0:.2f} points\n'.format(-a[min_row,3]+a[min_row,0]))

print('The highest daily gain in stock market history occured on {0}'.format(b[max_row]))
print('On that day, the stock market gained {0:.2f} points\n'.format(a[max_row,3]-a[max_row,0]))



# Assignment 3 Problem 2
# the most daily transaction volume and its date

print("Most Daily transaction volume: {0:.0f}".format(np.amax(a, axis=0)[5]))
print("Date of this transaction volume: {0}\n".format(b[np.argmax(a, axis=0)[5]]))


# Assignment 3 Problem 3
# monthly report for 2017-2018

current_month1 = '0'
current_month2 = '1'
open_total = 0
close_total = 0
volume_total = 0
days = 0

final_arr = np.array([])

for i in range(len(b)):
    if (b[i][0:4] == '2017' or b[i][0:4] == '2018'):
        if (b[i][5] == current_month1 and b[i][6] == current_month2):
            open_total += a[i][0]
            close_total += a[i][3]
            volume_total += a[i][5]
            days += 1
        
        else:
            if (current_month1 == '0' and current_month2 =='1'):
                final_arr = np.hstack((final_arr, np.array([open_total /days, close_total/days, volume_total/days, (close_total-open_total)/days, '01'])))
            else:
                final_arr = np.vstack((final_arr, np.array([open_total /days, close_total/days, volume_total/days, (close_total-open_total)/days, current_month1 + current_month2])))
            
            if (current_month1 == '1' and current_month2 == '2'):
                break
                
            days = 0
            open_total = 0
            close_total = 0
            volume_total = 0
            
            if current_month2 == '9':
                current_month1 = '1'
                current_month2 = '0'
            else:
                current_month2 = str(int(current_month2) + 1)
    

print('\n Monthly Average Report: \n\n{0}\n\n\n'.format(final_arr))

def find_months(low, high):
    l = final_arr
    return [x for x in l if float(x[3])>=low and float(x[3])<= high]
    
l = find_months(-15, 0)

print("The months that fall into the specified range are: \n")

for x in l:
    print (x)

    
    

    
# Assignment 3 Problem 4
# a yearly report which has annual average open price, close price, transaction volume, 
# and gain/loss from 1950 to 2018, and the most profitable year

current_year = ['1','9','5','0']
open_total = 0
close_total = 0
volume_total = 0
days = 0

arr = np.array([])

for i in range(len(b)):
    if (current_year[0] == b[i][0] and current_year[1] == b[i][1] and current_year[2] == b[i][2] and current_year[3] == b[i][3]):
        open_total += float(a[i][0])
        close_total+= float(a[i][3])
        volume_total += float(a[i][5])
        days += 1
        
    else:
        if (current_year[0] == '1' and current_year[1] == '9' and current_year[2] == '5' and current_year[3] == '0'):
            arr = np.hstack((arr, np.array([open_total/days, close_total/days, volume_total/days, (close_total-open_total)/days, '1950'])))
        
        else:
            arr = np.vstack((arr, np.array([open_total/days, close_total/days, volume_total/days, (close_total-open_total)/days, current_year[0] + current_year[1] + current_year[2] + current_year[3]])))
        
        
        if (current_year[2] == '1' and current_year[3] == '8'):
            break;
            
        days = 0
        volume_total = 0
        open_total = 0
        close_total = 0

        current_year = [b[i][0], b[i][1], b[i][2], b[i][3]]
        
        
print("\n\nYearly report: \n\n{0}\n\n".format(arr))
        
arr = arr.astype(float)
    
print("\n\n\nMost Profitable Year: {0:.0f}\n".format(arr[np.argmax(arr, axis=0)[3]][4]))
print("Average gain in that year: {0:.2f} points".format(np.amax(arr, axis=0)[3]))





# Assignment 3 Problem 5
# a every other five year report which has every five year average open price, close price, transaction volume,
# and gain/loss from 1950 to 2018, and the most profitable five year,


current_year = ['1','9','5','0']
open_total = 0
close_total = 0
volume_total = 0
days = 0

arr = np.array([])

for i in range(len(b)):
    if (b[i][2] == current_year[2] and int(b[i][3]) - int(current_year[3]) < 5):
        open_total += float(a[i][0])
        close_total+= float(a[i][3])
        volume_total += float(a[i][5])
        days += 1
        
    else:
        if (current_year[0] == '1' and current_year[1] == '9' and current_year[2] == '5' and current_year[3] == '0'):
            arr = np.hstack((arr, np.array([open_total/days, close_total/days, volume_total/days, (close_total-open_total)/days, '1950'])))
        
        else:
            arr = np.vstack((arr, np.array([open_total/days, close_total/days, volume_total/days, (close_total-open_total)/days, current_year[0] + current_year[1] + current_year[2] + current_year[3]])))
        
        
        if (current_year[2] == '1' and current_year[3] == '5'):
            break;
            
        days = 0
        volume_total = 0
        open_total = 0
        close_total = 0

        current_year = [b[i][0], b[i][1], b[i][2], b[i][3]]
        
        
print("\n\n Five-Year Report: \n\n{0}\n".format(arr))
        
arr = arr.astype(float)
    
print("\n\n\nMost Profitable 5 Year Period: {0:.0f}-{1:.0f}\n".format(arr[np.argmax(arr, axis=0)[3]][4], arr[np.argmax(arr, axis=0)[3]][4] + 4) )
print("Average gain in those years: {0: .2f} points".format(np.amax(arr, axis=0)[3]))
```
